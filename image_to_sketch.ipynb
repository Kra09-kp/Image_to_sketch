##Invert the grey image

Here, we invert our grey image to get more detail and make sketch very accurate

```python
invert_img = cv.bitwise_not(grey_img)
cv2_imshow(invert_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

##Blur inverted image

Here, we blur the inverted image with cv's GaussianBlur Function

```python
blur_img = cv.GaussianBlur(invert_img,(21,21),0)
cv2_imshow(blur_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

###invert blured image

here, we again invert our image after blurring to get more realistic sketch.

```python
invert_blur_img = cv.bitwise_not(blur_img)
cv2_imshow(invert_blur_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

##Pencil Sketch

Now we make pencil sketch from the blurred image by CV module.

```python
sketch = cv.divide(grey_img,invert_blur_img,scale=256.0)
cv2_imshow(sketch)
cv.waitKey(0)
cv.destroyAllWindows()
```
